```python
path_1 = r'C:\Users\User\Desktop\Quera_Shopping_Analysis\supermarket.csv'
df_supermarket = pd.read_csv(path_1)
df_supermarket
```

```python
list_customer = []
for costumer in df_supermarket['Customer Id'].unique():
    list_customer.append(list(df_supermarket['Product'][df_supermarket['Customer Id'] == costumer]))
dic_customer = {}
dic_customer['product'] = list_customer
```

```python
df_customer = pd.DataFrame(dic_customer)
```

```python
# First Question
Number_Products = len(df_supermarket['Product'].unique())
print(Number_Products)
```

```python
# Second Question
Total_Cost = len(df_supermarket['Product'])
Total_Day = len(df_supermarket['Date'].unique())
print("%.2f" % float(Total_Cost/Total_Day))
```

```python
# Third Question
Product_Consumption = {}
for product in df_supermarket['Product'].unique():
    Product_Consumption[product] = len(df_supermarket[df_supermarket['Product'] == product])
low_four_product_list = []
for i in range(4):
    key_list = list(Product_Consumption.keys())
    val_list = list(Product_Consumption.values())
    low_four_product_list.append(key_list[val_list.index(min(val_list))])
    del(Product_Consumption[key_list[val_list.index(min(val_list))]])
for element in low_four_product_list:
    if element != low_four_product_list[-1]:
        print(element,end=',')
    else:
        print(element)
```

```python
Basket_Customer = {}
for customer in df_supermarket['Customer Id'].unique():
    number = list(df_supermarket['Date'][df_supermarket['Customer Id'] == customer])
    date = []
    for element in number:
        if element[0:4] == '2020':
            date.append(element)
    set_date = set(date)      
    Basket_Customer[customer] = len(set_date)
list(Basket_Customer.items())[0:5]
```

```python
# Fourth Question  
Five_Most_Basket_Customer = []
for i in range(5):
    key_list = list(Basket_Customer.keys())
    val_list = list(Basket_Customer.values())
    Five_Most_Basket_Customer.append(key_list[val_list.index(max(val_list))])
    del(Basket_Customer[key_list[val_list.index(max(val_list))]])
for element in Five_Most_Basket_Customer:
    if element != Five_Most_Basket_Customer[-1]:
        print(element,end=',')
    else:
        print(element)
```

```python
df_supermarket['Date_changed'] = df_supermarket['Date']
df_supermarket['Date_changed'].astype('str')
df_supermarket['Date_changed'] = df_supermarket['Date_changed'].str.replace('-','')
df_supermarket['Date_changed'] = df_supermarket['Date_changed'].astype('str')
df_supermarket
```

```python
# Fifth Question
weekDays = ("Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday")
Day_list = []
for date in df_supermarket['Date_changed']:
    Day = weekDays[datetime(int(date[0:4]),int(date[4:6]),int(date[6:8])).weekday()]
    Day_list.append(Day)
df_supermarket['Weekday'] = Day_list
df_supermarket['Weekday'].value_counts().idxmax()
```

```python
df_supermarket = pd.get_dummies(df_supermarket,columns = ['Product'])
df_supermarket.columns = df_supermarket.columns.str.replace("Product_", "")
Product_List = []
for column in df_supermarket.columns:
    if column != 'Customer Id' and column != 'Date' and column != 'Date_changed' and column != 'Weekday':
        Product_List.append(column)
Product_List
```

```python
a = TransactionEncoder()
a_data = a.fit(list_customer).transform(list_customer)
df = pd.DataFrame(a_data,columns=a.columns_)
df = df.replace(False,0)
df = df.replace(True,1)
df
```

```python
df = apriori(df, min_support = 0.01, use_colnames = True, verbose = 1)
df
```

```python
# Sixth Question
support = {}
for row in df.index:
    support[df['itemsets'][row]] = df['support'][row]
Five_Most_Support = []
for i in range(5):
    key_list = list(support.keys())
    val_list = list(support.values())
    Five_Most_Support.append(key_list[val_list.index(max(val_list))])
    del(support[key_list[val_list.index(max(val_list))]])
for element in Five_Most_Support:
    if element != Five_Most_Support[-1]:
        print(element,end=',')
    else:
        print(element)
```

```python
df_ar = association_rules(df, metric = "confidence", min_threshold = 0.7)
df_ar
```

```python
# Seventh Question
confidence = {}
for row_2 in df_ar.index:
    confidence[str(df_ar['antecedents'][row_2]) + str('->') + str(df_ar['consequents'][row_2])] = df_ar['confidence'][row_2]
Two_Most_Confidence = []
for i in range(2):
    key_list = list(confidence.keys())
    val_list = list(confidence.values())
    Two_Most_Confidence.append(key_list[val_list.index(max(val_list))])
    del(confidence[key_list[val_list.index(max(val_list))]])
for element in Two_Most_Confidence:
    if element != Two_Most_Confidence[-1]:
        print(element,end='|')
    else:
        print(element)
```
